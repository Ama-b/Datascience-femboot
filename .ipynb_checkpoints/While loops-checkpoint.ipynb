## WHILE LOOP

```python
#while cond:
#    something to happen if true

x = 12
while x>2:
    print("It is true")
```

```python
x = 12
while x>2:
    print("It is true")
    x = x/2
```

```python
list=[]
number=20
while number>=0:
    number=float(input("Enter a number: "))
    list.append(number)
list.pop()
print(list)
```

```python
list=[]
while True:
    number=float(input("Enter a number: "))
    if number<0:
        break
    else:
    list.append(number)
print(list)
```
