```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model=LinearRegression()
```

```python
model
```

```python
boston=load_boston()
```

```python
print(boston.DESCR)
```

```python
X=boston.data
y=boston.target
```

## Age

```python
X_age=X[:,6]
```

```python
X_age=X_age.reshape(-1,1)
```

```python
X_age
```

```python
train_test_split(X_age,y)
```

```python
X_age_train,X_age_test,y_train,y_test=train_test_split(X_age,y,test_size=0.3)
```

```python
X_age_train.shape
```

```python
model.fit(X_age_train,y_train)
```

```python
model.score(X_age_train,y_train)
```

```python
model.score(X_age_test,y_test)
```
